```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('sales_data.xlsx')
df
```

```python
# Clean column names
df.columns = df.columns.str.strip()  # Remove leading/trailing spaces
```

```python
# Convert 'Date' to datetime
df['Date'] = pd.to_datetime(df['Date'])
df
```

```python
# Add 'Month' column
df['Month'] = df['Date'].dt.to_period('M')
df
```

```python
# -------- INSIGHTS -------- #

# Total revenue by region
revenue_by_region = df.groupby('Region')['Revenue'].sum().sort_values(ascending=False)
print("Revenue by Region:\n", revenue_by_region)
```

```python
# Top-selling products by units sold
top_products = df.groupby('Product')['Units Sold'].sum().sort_values(ascending=False)
print("\nTop Products:\n", top_products)
```

```python
# Monthly revenue trend
monthly_revenue = df.groupby('Month')['Revenue'].sum()
print("\nMonthly Revenue:\n", monthly_revenue)
```

```python
# -------- VISUALIZATIONS -------- #
```

```python
# Set seaborn theme
sns.set_theme(style="whitegrid")
```

```python
# 1. Bar plot - Revenue by Region
plt.figure(figsize=(6, 3))
sns.barplot(x=revenue_by_region.index, y=revenue_by_region.values, hue=revenue_by_region.index, palette='viridis', legend=False)
plt.title('Total Revenue by Region')
plt.xlabel('Region')
plt.ylabel('Revenue')
plt.tight_layout()
plt.show()
```

```python
# --- Visualization 2: Sales Distribution by Product ---
plt.figure(figsize=(5, 5))
plt.pie(top_products, labels=top_products.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette('pastel'))
plt.title('Sales Distribution by Product')
plt.axis('equal')  # Ensure pie chart is circular
plt.tight_layout()
plt.show()
```

```python
# --- Visualization 3: Monthly Revenue Line Chart ---
plt.figure(figsize=(7, 3))
sns.lineplot(x=monthly_revenue.index.astype(str), y=monthly_revenue.values, marker='o', color='green')
plt.title('Monthly Revenue Trend - NovaEdge Distributors Ltd.')
plt.xlabel('Month')
plt.ylabel('Total Revenue (₦)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
